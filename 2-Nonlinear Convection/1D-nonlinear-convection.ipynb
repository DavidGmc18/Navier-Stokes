```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output

nx = 201
dx = 10 / (nx - 1)

nt = 200
dt = 0.025

u = np.ones(nx)
u[int(0.5 / dx) : int(1 / dx + 1)] = 2

print("Total: ", np.trapezoid(u, dx=dx), "U", sep="")
```

```python
for n in range(nt):
    un = u.copy()
    for i in range(1, nx):
        u[i] = un[i] - un[i] * dt / dx * (un[i] - un[i-1])

    clear_output(True)
    plt.plot(np.linspace(0, 10, nx), u)
    plt.axis([0, 10, 1, 2.1])
    plt.gcf().set_size_inches(10, 4)
    plt.pause(0.01)
    print("Total: ", np.trapezoid(u, dx=dx), "U", sep="")
```
