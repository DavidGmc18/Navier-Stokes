```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output

nx = 41
dx = 2 / (nx - 1)

nt = 25
dt = 0.025

c = 1

u = np.ones(nx)
u[int(0.5 / dx) : int(1 / dx + 1)] = 2

print("Total: ", np.trapezoid(u, dx=dx), "U", sep="")
```

```python
for n in range(nt):
    un = u.copy()
    for i in range(1, nx):
        u[i] = un[i] - c * dt / dx * (un[i] - un[i-1])

    clear_output(True)
    plt.plot(np.linspace(0, 2, nx), u)
    plt.pause(0.01)
    print("Total: ", np.trapezoid(u, dx=dx), "U", sep="")
```
