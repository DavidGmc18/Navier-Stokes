```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output

nx = 41
dx = 2 / (nx - 1)

t = 0.25
sigma = 0.3

nu = 0.3

dt = sigma * dx**2 / nu
nt = int(t / dt)

u = np.ones(nx)
u[int(0.5 / dx):int(1 / dx + 1)] = 2

for n in range(nt):
    un = u.copy()
    for i in range(1, nx - 1):
        u[i] = un[i] + dt * nu * (un[i+1] - 2 * un[i] + un[i-1]) / dx**2

    u[0] = u[1]
    u[-1] = u[-2]

    clear_output(True)
    plt.plot(np.linspace(0, 2, nx), u)
    plt.axis([0, 2, 0, 2.1])
    plt.pause(0.01)

```
